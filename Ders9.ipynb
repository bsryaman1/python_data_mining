```python
df_cust["fea_2"].fillna(df_cust["fea_2"].mean(),inplace=True)
df_cust.isna().sum()
```

```python
y=df_cust["label"] #y -> output
X=df_cust.drop(["label","id"],axis=1) #x-> input
X
```

```python
from collections import Counter
```

```python
Counter(y)
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
sampler=RandomOverSampler()
X_new, y_new=sampler.fit_resample(X,y)
```

```python
Counter(y_new)
```

```python
X_new.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_new,y_new,test_size=0.2,random_state=42)
#her zaman model olusturulurken model train ve test olarak Ä±kÄ±ye bolunur
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
print(f"Train set accurarcy score={accuracy_score(y_train,model.predict(X_train))}")
print(f"Test set accurarcy score={accuracy_score(y_test,model.predict(X_test))}")
```

```python
tree.plot_tree(model)
```

```python
path=model.cost_complexity_pruning_path(X_train,y_train)
alphas=path["ccp_alphas"]
```

```python
accuracy_train,accuracy_test=[],[]
for i in alphas:
    model_prune = DecisionTreeClassifier(ccp_alpha=i)
    
    model_prune.fit(X_train, y_train)
    y_train_pred = model_prune.predict(X_train)
    y_test_pred = model_prune.predict(X_test)
    
    accuracy_train.append(accuracy_score(y_train,y_train_pred))
    accuracy_test.append(accuracy_score(y_test,y_test_pred))
```

```python
sns.lineplot(y=accuracy_train,x=alphas,label='EÄŸitim DoÄŸruluÄŸu')
sns.lineplot(y=accuracy_test,x=alphas,label='Test DoÄŸruluÄŸu')
plt.xlabel("Alfa DeÄŸerleri")
plt.ylabel("DoÄŸruluk")
plt.xticks()
plt.show()
```

```python
model_new=DecisionTreeClassifier(ccp_alpha=0.08)
model_new.fit(X_train,y_train)
```

```python
print(f"Train set accurarcy score={accuracy_score(y_train,model_new.predict(X_train))}")
print(f"Test set accurarcy score={accuracy_score(y_test,model_new.predict(X_test))}")
```

```python
tree.plot_tree(model_new)
```

```python
sns.heatmap(confusion_matrix(y_test,model_new.predict(X_test)),
           annot=True,fmt="g") 
#tahminler ve dogru tahmin hakkÄ±nda bu matrix bÄ±lgÄ± verÄ±r
#dogru ve yanlÄ±s tahmÄ±nlerÄ± verÄ±r-> confusipn matrix(karmasÄ±klÄ±k matrÄ±sÄ±)
#0-1 -> yanlÄ±s, 1-1 ve 0-0 dogru
```

```python
def rank_pokemon(pokemon_type):
    if pokemon_type in ["Grass","Fire","Water"]:
        return "Classic"
    elif pokemon_type=="Normal":
        return "Boring"
    else:
        return "TBD"
    
```

```python
pokemon.apply(rank_pokemon)
```

```python
df=pd.DataFrame(columns=fortune.columns)
#yenı dataframe olusturma. Columlar fortune ıle aynı
```

```python
#her sektorde en yuksek kazanca sahıp sirket;
for sector, data in sectors:
    highst_profit_in_sector=data.nlargest(1,"Profits")#ikinci parametre olarak istenileni ver
    df=df.append(highst_profit_in_sector) #burada df.append yerıne yenı versıyonda pd.concat kullan.
# sector ve data olarak iki nesne verilme sebebi nlargestta tuple donmesı.
    df
```

```python
week1=pd.read_csv("resources/Restaurant - Week 1 Sales.csv")
week2=pd.read_csv("resources/Restaurant - Week 2 Sales.csv")
foods=pd.read_csv("resources/Restaurant - Foods.csv")
customers = pd.read_csv("resources/Restaurant - Customers.csv")
```

```python
pd.concat([week1,week2],ignore_index=True) 
#iki df bırlestırırıldı
#ignore_index-> var olan indexleri ihmal eder ve kendı index olusturur.
```

```python
sales=pd.concat([week1,week2],keys=["Week1","Week2"])
sales
#multıindex yapıldı
```

```python
week1.merge(week2, how="inner", on="Customer ID") 
#inner ortak alan demek, on-> hangi alanı almak istiyoruz 
week1.merge(week2, how="inner", on="Customer ID", suffixes=("- Week1","- Week2")) 
# WEEK1 VE WEEK 2 INNER(ORTAKLARA BAKARAK) birlestir. Customer ID-> ortak alan
```

```python
#how{‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’}, default ‘inner’
#inner-> ortak sutunlar uzerınden bırlestırme yapılır
#outer-> inner gibi ortak alan kullanır ama karsılıgı olmayanları da getırı. Bırınde olup dıgerınde olamayanlar da gozukur.
#left outer-> solunda bulunan tüm satırları listeler sağdakileri null yapar
#right outer-> sağdakı tum satırları getırır.Sol null olur.
```

```python
week1[week1["Customer ID"]==155]
#Customer ID 155 olan indexlerin tüm bilgileri Week 1 için
```

```python
week2[week2["Customer ID"]==155]
#Week 2 için Customer ID=155 indexlerin bilgileri
```

```python
week1.merge(week2,how="inner", on="Customer ID",suffixes=("_Week1","Week2"))
#suffixes-> birleştitilmeyen alanların isimlendirirlmesi
```

```python
week1.merge(week2,how="inner", on=["Customer ID","Food ID"])
#Customer ve Food ID ye gore bırlestırme
```

```python
week1.merge(week2, how="outer", on="Customer ID", suffixes=("_Week1","_Week2"),indicator=True)
#indicator-> nasıl birleştirdiğini soyler
```

```python
result=week1.merge(week2, how="outer", on="Customer ID", suffixes=("_Week1","_Week2"),indicator=True)
result
```

```python
condition=result["_merge"].isin(["left_only","right_only"])
result[condition]
#birlestirme sekıllerınden sadece left ve right only leri getırır.
```

```python
week2.merge(foods, how="left", on="Food ID").drop("Food ID", axis="columns")
#week 2 food ıle merge edılır.
#Food ID columnlardan kaldırırlır.
```

```python
week2.merge(customers, how="left", left_on="Customer ID", right_on="ID") 
#Customer df si week 2 ile merge edilir. sol var diye sağ vermemiz lazım
pd.merge(week2, customers, how="left", left_on="Customer ID", right_on="ID") 
# ikisi aynı sadece burda hem sağ hem de solu vermemiz lazım 
```

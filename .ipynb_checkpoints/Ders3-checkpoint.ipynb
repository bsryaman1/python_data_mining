```python
emp_df=pd.read_csv("resources/employees.csv",parse_dates=["Start Date","Last Login Time"])
#datetime veritipine çevirmek için ilk yol bu.
#date time a çevirmek için ikinci yol ;
#emp_df["Start Date"]=pd.to_datetime(emp_df["Start Date"])
#ikisiyle de aynı sonuç çıkar.

emp_df["Senior Management"]=emp_df["Senior Management"].astype(bool)
emp_df["Gender"]=emp_df["Gender"].astype("category")
#bu şekilde doğrudan çevirme sağlanır.
emp_df.info()

```

```python
emp_df[emp_df["Gender"]=="Male"] #Genderı Male olanları getir demek
```

```python
cond=emp_df["Team"]=="Finance"
emp_df[cond]
#üstekinin aynısı ama daha yardımcı
```

```python
emp_df[emp_df["Senior Management"]] #Direkt True olanları getirir.
```

```python
emp_df[emp_df["Team"] != "Finance"] # eşit olmayanları getirir.
```

```python
emp_df[emp_df["Salary"] > 50000] # 50 binden büyük olanları getirir
```

```python
emp_df[emp_df["Start Date"] > "1985-01-01" ] # bu tarihten sonra gelenleri getirir
```

```python
emp_df[(emp_df["Team"]=="Marketing")& (emp_df["Start Date"]>"1990-01-01")]
# ve için & kullanılır. Aynı işlem için diğer yol;
cond1=emp_df["Team"]=="Marketing"
cond2=emp_df["Start Date"]> "1990-01-01"
emp_df[cond1 & cond2]
#bu yöntem daha iyi
```

```python
emp_df[(emp_df["Team"] == "Marketing") | (emp_df["Start Date"] > "1990-01-01")]  
# veya için | dik çizgi kullanırız
```

```python
cond1 = emp_df["First Name"] == "Robert"
cond2 = emp_df["Start Date"] > "2016-06-01"
cond3 = emp_df["Team"] == "Client Services"

emp_df[cond1 & cond3 | cond2 ] # ve, veya böyle kullanılır 
```

```python
emp_df[emp_df["Team"].isin(["Marketing","Finance","Client Services"])] 
#Team alanı Marketing, Fiance ve Client Services olanlar
```

```python
emp_df[emp_df["Team"].isnull()] #null olanları
```

```python
emp_df[emp_df["Team"].notnull()] #null olmayanları 
```

```python
emp_df[emp_df["Salary"].between(40000,50000,inclusive='both')] #40000-50000 arası geliri olanları getirir
#inclusive both dahil demek. neither olursa 40000 ve 50000 dahil olmaz
# right = sağ tarafı dahil et sol tarafı etme
# default hali both olur 
```

```python
emp_df[emp_df["Start Date"].between("1985-01-01","2020-01-01")] 
#bu aralık arası tarihte işe başlayanları getirir.
```

```python
emp_df[emp_df["Last Login Time"].between("08:30 AM", "12:00 PM")]
#bu iki saatin arasında giriş yapanlar 
```

```python
emp_df.sort_values("First Name",inplace=True) #first name göre sıralar
emp_df
```

```python
emp_df["First Name"].duplicated() #tekararlanan first nameları getirir.
emp_df
```

```python
emp_df["First Name"].duplicated(keep="last") #sonuncuyu saklayarak getirir.
emp_df
```

```python
emp_df["First Name"].duplicated(keep=False) #hiçbirini saklmadan getirir.
emp_df
```

```python
emp_df[~emp_df["First Name"].duplicated(keep=False)] #tekararlanmayan isimleri getirir.
emp_df
```

```python
emp_df.drop_duplicates() 
#tekraralanan kayıtları siler. Hepsi tam aynı olanlar.
```

```python
emp_df.drop_duplicates(subset=["First Name"])
#aynı isme sahip olanlardan ilki hariç siler
```

```python
emp_df.drop_duplicates(subset=["First Name"],keep=False)
#ilkini tutmadan hepsini siler
#keep ile hangisi tutulacaksa o seçilir.
```

```python
emp_df.drop_duplicates(subset=["First Name","Team"]) #istediğin kadar verilir.
#aynı isimden veya aynı kayıttan aynı olanların ilkini tutarak siler.
```

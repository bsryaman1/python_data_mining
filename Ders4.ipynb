```python
def ekle(number):
    return str(number) + "Milion $" 
#str ile seri için çevrirlir.
```

```python
jb_df["Bond_Actor_Salary"].apply(ekle)
#ekle fonkdiyonu apply ile tüm Bond_Actor_Salary columnuna uygulanır.
```

```python
def iyi_film(a):
    aktor = a[1]
    butce = a[4]
    
    if aktor == "Pierce Brosnan":
        return "En iyi"
    elif aktor =="Roger More" and butce > 40:
        return "Eğlecili"
    else:
        return "Fikrim yok"
    # a ile column numarası verilerek film değerlendirmesi yapılır.
```

```python
jb_df.apply(iyi_film, axis=1)
#axis=1 ile tüm indexlere iyi_film fonsiyonu apply ile uygulanır.
```

```python
df_city=pd.read_csv("resources/chicago.csv")
df_city.head()
```

```python
df_city.info() # bellekte ne kadar yer tuttuğunu görebiliriz
```

```python
df_city["Position Title"].str.title() #Kelimelerin baş harfını buyuk gerısını kucuk yapar.
#Burada seri oldugu ıcın fonksıyonları kullanmadan once str ıle serıye uygulanması saglanır.
```

```python
df_city["Position Title"].str.upper() #Tum harflerı buyuk yapar.
```

```python
df_city["Position Title"].str.lower() #Tum harflerı kucuk yapar.
```

```python
df_city["Position Title"].str.len() #uzunlugunu karkater sayısı olarak verır.
```

```python
df_city.dropna(how="all",inplace=True)
#satırın hepsi NaN ise siler.
```

```python
df_city["Department"].str.replace("MGMNT","MANAGEMENT") 
# department alanındaki MGMNT yazılan yerlere MANAGMENT yazar
```

```python
df_city["Employee Annual Salary"] = df_city["Employee Annual Salary"].str.replace("$","",regex=False).astype(float)
df_city
#Employee Annual Salary columndaki $ kaldırıldı ve floata çevrildi.
#str database olduğu için ve regex Regular Expressiondan geldıgı ıcın yazıldı.
```

```python
df_city[df_city["Department"].str.startswith("WATER")]
#WATER ile başlayan kelimeleri getirir Departmentte
```

```python
df_city[df_city["Department"].str.endswith("WATER")] 
# WATER ile biten kelimeler 
```

```python
df_city[df_city["Department"].str.contains("WATER")] # içinde WATER kelimesi olan kelimeler 
```

```python
df_city["Name"].str.lstrip() # String'in sol tarafındaki boşluğu siler
```

```python
df_city["Name"].str.rstrip() # String'in sağ tarafındaki boşluğu siler
```

```python
df_city["Name"].str.strip() 
# String'in hem sağ hem de sol tarafındaki boşluğu siler(kelimelerin arasındaki boşlukları silmiyor)
```

```python
df_city["Name"].str.split(",") 
#seri döndürür. split içine yazılan değerlere göre ayrır veya default haliyle boşlukları siler.
```

```python
df_city["Name"].str.split(",").str.get(0).str.title().value_counts().head() 
#en çok tekrar edeen ilk 5 değeri virgüle göre ayırark getirir.
# str.get(0) -> ilk kelimeyi ayırıyor
# str.title -> kelimelerin baş harfini büyük yapar
# value.counts -> bir kelimenin kaç defa kaullandığını gösterir
```
